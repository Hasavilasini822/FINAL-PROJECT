```python
#include <SoftwareSerial.h>
#include <Wire.h>
#include <Adafruit_GFX.h>
#include <Adafruit_SSD1306.h>
 
#define SCREEN_WIDTH 128    // OLED display width, in pixels
#define SCREEN_HEIGHT 64    // OLED display height, in pixels
#define OLED_RESET -1       // Reset pin # (or -1 if sharing Arduino reset pin)
Adafruit_SSD1306 display(SCREEN_WIDTH, SCREEN_HEIGHT, &Wire, OLED_RESET);

SoftwareSerial mod(2,3); 
 
void setup() {
  Serial.begin(9600);
  mod.begin(9600);
  
  display.begin(SSD1306_SWITCHCAPVCC, 0x3C); //initialize with the I2C addr 0x3C (128x64)
  delay(500);
  display.clearDisplay();
  display.setCursor(10, 15);
  display.setTextSize(1);
  display.setTextColor(WHITE);
  display.println("7 in 1 soil Sensor");
  display.setCursor(25, 35);
  display.setTextSize(1);
  display.print("Initializing");
  display.display();
  delay(1000);
}
 
void loop() {
  float val1 , val2 , val3 , val4 , val5 , val6 , val7;
  val1 = Nitrogen_Calculation();
  val2 = Phosphorus_Calculation();
  val3 = potassium_Calculation();
  val4 = ph_cal();
  val5 = temp_Calculation();
  val6= EC_Calculation();
  val7 = moisture_cal();
  
  
  Serial.print("Nitrogen: ");
  Serial.print(val1);
  Serial.println(" mg/kg");
  Serial.print("Phosphorous: ");
  Serial.print(val2);
  Serial.println(" mg/kg");
  Serial.print("Potassium: ");
  Serial.print(val3);
  Serial.println(" mg/kg");
  Serial.print("pH: ");
  Serial.print(val4);
  Serial.println(" ph");
  Serial.print("temp: ");
  Serial.print(val5);
  Serial.println("  'c");
  Serial.print("ec: ");
  Serial.print(val6);
  Serial.println("  %");

  Serial.print("moisture : ");
  Serial.print(val7);
  Serial.println("  %");
  
 
 
  display.clearDisplay();
  
 
  display.setTextSize(1);
  display.setCursor(10, 0);
  display.print("N: ");
  display.print(val1);
  display.setTextSize(1);
  display.print(" mg/kg");
 
  display.setTextSize(1);
  display.setCursor(10,9);
  display.print("P: ");
  display.print(val2);
  display.setTextSize(1);
  display.print(" mg/kg");
 
  display.setTextSize(1);
  display.setCursor(10,18);
  display.print("K: ");
  display.print(val3);
  display.setTextSize(1);
  display.print(" mg/kg");

  display.setTextSize(1);
  display.setCursor(10, 27);
  display.print("pH: ");
  display.print(val4);
  display.setTextSize(1);
  display.print(" ph");

  display.setTextSize(1);
  display.setCursor(10,36);
  display.print("Temp: ");
  display.print(val5);
  display.setTextSize(1);
  // display.print(" C");

  display.setTextSize(1);
  display.setCursor(10,45);
  display.print("EC: ");
  display.print(val7);
  display.setTextSize(1);
  // display.print(" %");

  // display.setTextSize(1);    ////////////
  // display.setCursor(10,45);
  // display.print("mois : ");
  // // display.print(val6);
  // display.setTextSize(1);
  // display.print(" %");      //////////////////

  

 
  display.display();
}
 
//////////////////////////////////////

float  ph_cal(){
  byte queryData[] = {0x01, 0x03, 0x00, 0x06, 0x00, 0x01, 0x64, 0x0b};
  byte phvalue[7];
  mod.write(queryData, sizeof(queryData));  // Send the query data to the NPK sensor
  delay(500);
  if (mod.available() >= sizeof(phvalue)) {  // Check if there are enough bytes available to read
    mod.readBytes(phvalue, sizeof(phvalue));
    float result = (float)((phvalue[3] << 8) | phvalue[4]) ;
     return  (result/100);
  }
  else { return 0;}

}

float  moisture_cal()
{
  byte queryData[] = {0x01, 0x03, 0x00, 0x12, 0x00, 0x01, 0x24, 0x0f};
  byte moisturevlaue[7];
  mod.write(queryData, sizeof(queryData));  // Send the query data to the NPK sensor
  delay(500);  // Wait for 1 second

  if (mod.available() >= sizeof(moisturevlaue)) {  // Check if there are enough bytes available to read
    mod.readBytes(moisturevlaue, sizeof(moisturevlaue));  // Read the received data into the receivedData array
    float result =(float)((moisturevlaue[3] << 8) | moisturevlaue[4]) ;
     return  (result/10);
  }
  else
  {
    return 0;
  }

}

float  EC_Calculation()
{
  byte queryData[] = {0x01, 0x03, 0x00, 0x15, 0x00, 0x01, 0x95, 0xce};
  byte ECvalue[7];
  mod.write(queryData, sizeof(queryData));  // Send the query data to the NPK sensor
  delay(500);  // Wait for 1 second

  if (mod.available() >= sizeof(ECvalue)) {  // Check if there are enough bytes available to read
    mod.readBytes(ECvalue, sizeof(ECvalue));  // Read the received data into the receivedData array
    float result = (float)((ECvalue[3] << 8) | ECvalue[4]) ;
     return  (result/10);
  }
  else
  {
    return 0;
  }
}

float  temp_Calculation()
{
  byte queryData[] = {0x01, 0x03, 0x00, 0x12, 0x00, 0x02, 0x64, 0x0e};
  byte tempvalue[9];
  mod.write(queryData, sizeof(queryData));  // Send the query data to the NPK sensor
  delay(500);  // Wait for 1 second

  if (mod.available() >= sizeof(tempvalue)) {  // Check if there are enough bytes available to read
    mod.readBytes(tempvalue, sizeof(tempvalue));  // Read the received data into the receivedData array
    float result = (float)((tempvalue[5] << 8) | tempvalue[6]) ;
     return  (result/10);
  }
  else
  {
    return 0;
  }

}

float  humidity_Calculation()
{
  byte queryData[] = {0x01, 0x03, 0x00, 0x12, 0x00, 0x02, 0x64, 0x0e};
  byte humidityvalue[9];
  mod.write(queryData, sizeof(queryData));  // Send the query data to the NPK sensor
  delay(500);  // Wait for 1 second

  if (mod.available() >= sizeof(humidityvalue)) {  // Check if there are enough bytes available to read
    mod.readBytes(humidityvalue, sizeof(humidityvalue));  // Read the received data into the receivedData array
    float result = (float)((humidityvalue[3] << 8) | humidityvalue[4]) ;
     return  (result/10);
  }
  else
  {
    return 0;
  }

}

float  Nitrogen_Calculation()
{
  byte queryData[] = {0x01, 0x03, 0x00, 0x1e, 0x00, 0x01, 0xe4, 0x0c};
  byte nitrovalues[7];
  mod.write(queryData, sizeof(queryData));  // Send the query data to the NPK sensor
  delay(500);  // Wait for 1 second

  if (mod.available() >= sizeof(nitrovalues)) {  // Check if there are enough bytes available to read
    mod.readBytes(nitrovalues, sizeof(nitrovalues));  // Read the received data into the receivedData array
    float result = (float)((nitrovalues[3] << 8) | nitrovalues[4]) ;
     return  result;
  }
  else
  {
    return 0;
  }

}


float  Phosphorus_Calculation()
{
  byte queryData[] = {0x01, 0x03, 0x00, 0x1f, 0x00, 0x01, 0xb5, 0xcc};
  byte phosphorusvalues[7];
  mod.write(queryData, sizeof(queryData));  // Send the query data to the NPK sensor
  delay(500);  // Wait for 1 second

  
  
  if (mod.available() >= sizeof(phosphorusvalues)) {  // Check if there are enough bytes available to read
    mod.readBytes(phosphorusvalues, sizeof(phosphorusvalues));  // Read the received data into the receivedData array
    float result = (float)((phosphorusvalues[3] << 8) | phosphorusvalues[4]) ;
    return  result ;
  }
  else
  {
    return 0;
  }

}


float  potassium_Calculation()
{
  byte queryData[] = {0x01, 0x03, 0x00, 0x20, 0x00, 0x01, 0x85, 0xc0};
  byte potassiumValues[7];
  mod.write(queryData, sizeof(queryData));  // Send the query data to the NPK sensor
  delay(500);  // Wait for 1 second

  if (mod.available() >= sizeof(potassiumValues)) {  // Check if there are enough bytes available to read
    mod.readBytes(potassiumValues, sizeof(potassiumValues));  // Read the received data into the receivedData array
    float result = (float)((potassiumValues[3] << 8) | potassiumValues[4]) ;
     return  result ;
  }
  else
  {
    return 0;
  }

}
```
